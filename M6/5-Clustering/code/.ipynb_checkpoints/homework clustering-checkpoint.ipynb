```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
s_geo = pd.read_excel("../data/geo.xlsx")
s_comment    = pd.read_excel("../data/geo_comment.xlsx")
```

```python
X = s_geo[['x','y']]
С = s_geo[['comment_class']]
```

```python
X['y']
```

```python
plt.scatter(X['x'], X['y'], cmap='autumn', s=60)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
plt.scatter(X_scaled[:,0], X_scaled[:,1], cmap='autumn', s=60)
```

```python
X_neg = X_scaled[С['comment_class']==-1]
X_pos = X_scaled[С['comment_class']==1]
```

```python
fig, axs = plt.subplots(1,3)
fig.set_figheight(6)
fig.set_figwidth(12)

h1 = axs[0].scatter(X_scaled[:,0], X_scaled[:,1],c=С['comment_class'], cmap='autumn', s=60)
axs[0].set_title('Common')
h1 = axs[1].scatter(X_neg[:,0], X_neg[:,1], cmap='autumn', s=60)
axs[1].set_title('Negativity')
h1 = axs[2].scatter(X_pos[:,0], X_pos[:,1], cmap='autumn', s=60)
axs[2].set_title('Positive')
                        
plt.show()


#plt.scatter(X_scaled[:,0], X_scaled[:,1],c=С['comment_class'], cmap='autumn', s=60)
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, AffinityPropagation
from sklearn.metrics import adjusted_rand_score, silhouette_score
```

```python
k_inertia = []
ks = range(1,20)

for k in ks:
    clf_kmeans = KMeans(n_clusters=k)
    clusters_kmeans = clf_kmeans.fit_predict(X_scaled, )
    k_inertia.append(clf_kmeans.inertia_)
```

```python
plt.plot(ks, k_inertia)
plt.plot(ks, k_inertia ,'ro')
```

```python
diff = np.diff(k_inertia)
plt.plot(ks[1:], diff)
plt.plot(ks[1:], diff,'ro')
```

### По графику вижу перелом с 4 на 5 кластер.  

```python
diff_r = diff[1:] / diff[:-1]
plt.plot(ks[1:-1], diff_r)
plt.plot(ks[1:-1], diff_r,'ro')
```

### По этому графику вижу что минимумы при 3 и 16 кластерах

## Строим графики и оценки для 3, 4, 5 и 16 кластерах

```python
clf_kmeans = KMeans(n_clusters=3)
clusters_kmeans = clf_kmeans.fit_predict(X_scaled, )

fig, ax = plt.subplots()
ax.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters_kmeans, cmap='autumn', s=60)
ax.set_title('Оценка: {:.2f}'.format(silhouette_score(X_scaled, clusters_kmeans)))

plt.show()
```

```python
clf_kmeans = KMeans(n_clusters=4)
clusters_kmeans = clf_kmeans.fit_predict(X_scaled, )

fig, ax = plt.subplots()
ax.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters_kmeans, cmap='autumn', s=60)
ax.set_title('Оценка: {:.2f}'.format(silhouette_score(X_scaled, clusters_kmeans)))

plt.show()
```

```python
clf_kmeans = KMeans(n_clusters=5)
clusters_kmeans = clf_kmeans.fit_predict(X_scaled, )

fig, ax = plt.subplots()
ax.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters_kmeans, cmap='autumn', s=60)
ax.set_title('Оценка: {:.2f}'.format(silhouette_score(X_scaled, clusters_kmeans)))

plt.show()
```

```python
clf_kmeans = KMeans(n_clusters=16)
clusters_kmeans = clf_kmeans.fit_predict(X_scaled, )

fig, ax = plt.subplots()
ax.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters_kmeans, cmap='autumn', s=60)
ax.set_title('Оценка: {:.2f}'.format(silhouette_score(X_scaled, clusters_kmeans)))

plt.show()
```

### По оценкам можно сказать, что 0,57 максимальная оценка на 16 кластерах.
