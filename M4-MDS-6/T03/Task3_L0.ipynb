#### Придумайте матрицы, которые обладают следующими свойствами:

1. диагональная, ортогональная, но не единичная(!):
2. верхнетреугольная, ортогональная:
3. нижнетреугольная, ортогональная:
4. матрица, у которой определитель = 1, и все элементы не равны 0.

P.S. Во всех пунктах ответом должна быть ненулевая матрица размерности не менее 2!


```python
# 1. диагональная, ортогональная, но не единичная(!):
# поворот на 90 градусов.
#  sin от pi=0.Таким образом получается диагональная  ортогональная матрица 
x1 = np.array([np.cos(np.pi),0,0,np.cos(np.pi)])
x1 = x1.reshape(2,2)
print(x1)
print(x1.T)
print(np.linalg.inv(x1))
print(np.dot(x1,x1.T))
print(np.linalg.det(x1))
```

```python
# 2. верхнетреугольная, ортогональная:
# 3. нижнетреугольная, ортогональная:

# То-ли я не понял вопрос, то-ли вопрос с "подвохом"
# верхнетреугольная матрица - это, где элементы выше основной диоганали не нулевые, а под основной диагональю нулевые.
# как известно, ортогональная матрица, делает только вращения, а площадь под фигурой описываемой векторами остается неизменой.
# если элементы верхнего угла примут не нулевее начения, а элементы нижнего угла примут нулевые значения,
# то один угол сместиться, а другой останется на месте и площадь фигуры неминуемо измениться.
# Таким образом можно предположить, что нижнетреугольной или верхнетреугольной ортогональной матрицы не существует
# P.S. Я могу ошибаться, тогда я не понимаю, как придумать такую матрицу
```

```python
# 4. матрица, у которой определитель = 1, и все элементы не равны 0.
# матрица у кторой определитель = 0 - это и есть ортогональная матрица
# пример трехмерной матрицы поворота на 30 градусов во всех трех плоскостях
alpha = np.pi/6
beta = np.pi/6
gamma = np.pi/6 
x2 = np.array([np.cos(alpha)*np.cos(gamma)-np.sin(alpha)*np.sin(beta)*np.sin(gamma),
               -1*np.sin(alpha)*np.cos(beta),
               -1*np.cos(alpha)*np.sin(gamma)-np.sin(alpha)*np.sin(beta)*np.cos(gamma),
               np.cos(alpha)*np.sin(beta)*np.sin(gamma)+np.sin(alpha)*np.cos(gamma),
               np.cos(alpha)*np.cos(beta),
               np.cos(alpha)*np.sin(beta)*np.cos(gamma)-np.sin(alpha)*np.sin(gamma),
               np.cos(beta)*np.sin(gamma),
               -1*np.sin(beta),
               np.cos(beta)*np.cos(gamma)]).reshape(3,3)

print(x2)
# Проверки:
print(x2.T)
print(np.linalg.inv(x2))
# Транспонированая = обратной
print(np.dot(x2,x2.T))
# Произведение матрицы на транспонированную = едиичная матрица
print(np.linalg.det(x2))
# Определитель = 1
```
